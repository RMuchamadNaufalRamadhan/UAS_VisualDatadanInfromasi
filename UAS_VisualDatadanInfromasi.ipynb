```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kebutuhan_per_bulan_2.csv')
```

```python
data_filtered = df [df['Propinsi']=='ACEH']
data_filtered
```

```python
x = data_filtered['kebutuhan']
y = data_filtered['tahun']
plt.scatter(x, y)

plt.xlabel('Kebutuhan')
plt.ylabel('tahun')

plt.title('DATA KEBUTUHAN PER TAHUN DI ACEH')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Membaca dataset dari file CSV
df = pd.read_csv('kebutuhan_per_bulan_2.csv')

# Mengambil subset acak dari data
subset_size = 10  # Jumlah subset yang akan ditampilkan
subset_df = df.sample(n=subset_size, random_state=17)

# Memilih kolom yang akan digunakan sebagai nilai x dan y
x = subset_df['kebutuhan']
y = subset_df['Propinsi']
plt.figure(figsize=(10, 8))
# Membuat scatter plot
plt.scatter(x, y)

# Memberi label sumbu x dan y
plt.xlabel('Kebutuhan')
plt.ylabel('Propinsi')

# Memberi judul grafik
plt.title('Scatter Plot Secara Acak')

# Menampilkan grafik
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from CSV
data = pd.read_csv('profile_pelanggan.csv', header=None, names=['Kebutuhan', 'Propinsi'])

# Create scatter plot
plt.scatter(data['Propinsi'], data['Kebutuhan'])
plt.xlabel('Propinsi')
plt.ylabel('Kebutuhan')
plt.title('Scatter Plot Kebutuhan per Propinsi')
plt.xticks(rotation=90)
plt.show()
```
